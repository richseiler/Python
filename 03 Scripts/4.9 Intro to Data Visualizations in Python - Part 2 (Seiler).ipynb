```python
#import libraries

import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
# Create a path

path = r'C:\Users\Owner\Documents\Instacart Basket Analysis'
```

```python
# Import customers_merged.pkl

df = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'customers_merged.pkl'))
```

# Histogram

```python
# histogram of the 'order_hour_of_day' column

hist = df['order_hour_of_day'].plot.hist(bins = 24)
```

```python
df['order_hour_of_day'].value_counts(dropna=False)
```

Conclusion: The majority of the orders happen between 8am and 6pm.

# Bar Chart

```python
# bar chart for the 'loyalty_flag' column

bar = df['loyalty_flag'].value_counts().sort_index().plot.bar()
```

Conclusion: New customers (max order has fewer than 10 items)have a lower average because they're just trying it out for the first time. But regular customers (between 10 and 40 max items) and loyal customers (max order over 40 items) are what drives the business.

# Line Graph

```python
# Creating a sample

dev = np.random.rand(len(df)) <= 0.7
```

```python
# store 70% of the sample in the dataframe 'big'

big = df[dev]
```

```python
# store 30% of the sample in the dataframe 'small'

small = df[~dev]
```

```python
# new dataframe without unnecessary columns

df_2 = small[['order_hour_of_day','prices']]
```

```python
# line chart for 'prices' and 'order_hour_of_day' columns

line = sns.lineplot(data = df_2, x = 'order_hour_of_day', y = 'prices')
```

```python
# new dataframe without unnecessary columns

df_3 = small[['n_dependants','Age']]
```

```python
# line chart for 'age' and 'dependants_number' columns

line_2 = sns.lineplot(data = df_3, x = 'Age', y = 'n_dependants')
```

Conclusion: The number of dependants is between 1.4 and 1.6 no matter the age.

```python
pd.options.display.max_rows = None
```

```python
df_3[['Age', 'n_dependants']].groupby(['Age', 'n_dependants']).value_counts()
```

```python
# new dataframe without unnecessary columns

df_4 = small[['Income','Age']]
```

```python
# connection between age and income

sctp = sns.scatterplot(x = 'Age', y = 'Income', data = df_4)
```

Conclusion: Most of the data is at 200,000 dollars or less. After age 40, the data starts to trend up above 200,000 dollars at a greater rate.
