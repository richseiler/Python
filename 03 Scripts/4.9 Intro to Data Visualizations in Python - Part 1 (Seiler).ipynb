## Step 5. Complete the fundamental data quality and consistency checks youâ€™ve learned throughout this Achievement; for example, check for and address missing values and duplicates, and convert any mixed-type data.

```python
# Check data types

df_customers.dtypes
```

```python
# change userid from int64 to string

df_customers['user_id'] = df_customers['user_id'].astype('str')
```

```python
# check that the change worked

print(df_customers['user_id'].dtypes)
```

```python
# Basic statistics

df_customers.describe()
```

```python
# check the data frame for null values in each column

df_customers.isnull().sum()
```

```python
# check for columns that have mixed data types that may make analysis in the future complicated

for col in df_customers.columns.tolist():
    weird = (df_customers[[col]].applymap(type) != df_customers[[col]].iloc[0].apply(type)).any(axis=1)
    if len (df_customers[weird]) > 0:
        print (col)
```

```python
# Create new df where the First Name column is blank

df_nan = df_customers[df_customers["First Name"].isnull() == True]
```

```python
df_nan
```

A lot of rows are missing the first name, but it shouldn't be deleted because it shouldn't have an impact on the data if we keep it.

```python
# Changing First Name to a string value

df_customers['First Name'] = df_customers['First Name'].astype('str')
```

```python
# Check for missing values again

df_customers.isnull().sum()
```

Now that First_Name variable is classified to a string variable, it no longer is considered a null value anymore.
